## Data

The ideal dataset(s) we would like to use would include variables such as but not limited to unique animal ID, species, breed, primary color, age, intake date/time, intake condition, outcome date/time, outcome type, and found location. This dataset would ideally have thousands of animal records in order for us to have statistical significance. The data should be collected by an official source, likely an animal shelter, and be recorded over a multi-year period. The data should be saved in a CSV format for us to efficiently process for analysis.

We have already identified a potential dataset that fits this description. The Austin Animal Shelter has saved data from the years 2013 through 2025 for over 170,000 unique animals (outcome IDs). The dataset includes a lot of our desired variables, such as species, breed, color, age, intake data/time, intake condition, outcome data/time, outcome type, and found location. Thankfully, the data is considered open to the public as part of [Austin's Open Data Initiative](https://data.austintexas.gov/). The dataset can be found in two parts at the following URLs and will have to be joined using unique IDs:
